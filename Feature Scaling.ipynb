```python
import seaborn as sns 
import pandas as pd 
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Scalings brings all the features on same scale
```

```python
# Why scaling >> to reduce computation time
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
# sandardization >> Zsocre = x-mu/sigma
```

```python
mean = np.mean(df['total_bill'])
```

```python
std = np.std(df['total_bill'])
```

```python
std_data = []
for i in list(df['total_bill']):
    zscore = (i-mean)/std
    std_data.append(zscore)
```

```python
std_data
```

```python
#  scaling is optional because it doesnt change the distribution of data
sns.displot(df['total_bill'])
```

```python
sns.distplot(std_data)
```
